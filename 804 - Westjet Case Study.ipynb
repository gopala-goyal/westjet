```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', 200)
```

```python
df_westjet = pd.read_excel('WestJet Data 4-12-22.xlsx')
```

```python
print(df_westjet.info())
print(df_westjet.shape)
print(df_westjet.shape[0])
print(df_westjet.describe())
print(df_westjet.isnull().sum())
```

### Remove Empty Columns and Duplicate Rows

```python
empty_cols = [col for col in df_westjet.columns if df_westjet[col].isnull().all()]
df_westjet.drop(empty_cols,axis=1,inplace=True)
df_westjet.drop_duplicates(keep = 'first', inplace = True, ignore_index = True)
```

### Change Column Format and Clean data

```python
df_westjet.Date = pd.to_datetime(df_westjet.Date)
df_westjet.OTPArrivalBucket = df_westjet.OTPArrivalBucket.str.replace('min','').str.strip()
df_westjet.OTPArrivalBucket = df_westjet.OTPArrivalBucket.str.replace(' to ','-')
df_westjet.OTPArrivalBucket = df_westjet.OTPArrivalBucket.str.replace('More than 5 hours','300')
df_westjet.OTPArrivalBucket = df_westjet.OTPArrivalBucket.str.replace('Null',"")
df_westjet.OTPArrivalBucket = df_westjet.OTPArrivalBucket.str.replace('Early',"")
df_westjet.OTPDepBucket = df_westjet.OTPDepBucket.str.replace('min','').str.strip()
df_westjet.OTPDepBucket = df_westjet.OTPDepBucket.str.replace(' to ','-')
df_westjet.OTPDepBucket = df_westjet.OTPDepBucket.str.replace('More than 5 hours','300')
df_westjet.LFBucket = df_westjet.LFBucket.str.replace(' to ','-')
```
